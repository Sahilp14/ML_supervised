```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=2, n_redundant=0)
```

```python
X
```

```python
sns.scatterplot(pd.DataFrame(X)[0], pd.DataFrame(X)[1], hue=y)
```

```python
## Split the dataset into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='linear')
```

```python
svc.fit(X_train, y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
rbf = SVC(kernel='rbf')
```

```python
rbf.fit(X_train, y_train)
```

```python
y_pred1 = rbf.predict(X_test)
```

```python
print(classification_report(y_test, y_pred1))
print(confusion_matrix(y_test, y_pred1))
```

```python
poly = SVC(kernel='poly')
poly.fit(X_train, y_train)
y_pred2 = poly.predict(X_test)
print(classification_report(y_test, y_pred2))
print(confusion_matrix(y_test, y_pred2))
```

```python
sigm = SVC(kernel='sigmoid')
sigm.fit(X_train, y_train)

y_pred3 = sigm.predict(X_test)
print(classification_report(y_test, y_pred3))
print(confusion_matrix(y_test, y_pred3))
```

## HyperParameter tuning with SVC

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'C':[0.1,1,10,100,1000],
    'gamma':[1,0.1,0.01,0.001,0.0001],
    'kernel':['rbf']
}
```

```python
grid = GridSearchCV(SVC(), param_grid=param_grid, cv=5, refit=True,verbose=3)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred4 = grid.predict(X_test)
print(classification_report(y_test, y_pred4))
print(confusion_matrix(y_test, y_pred4))
```

```python
import seaborn as sns
df = sns.load_dataset('tips')
```

```python
X = df[['tip', 'sex', 'smoker', 'day', 'time', 'size']]
y = df['total_bill']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
print(X.shape)
print(len(y))
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le1 = LabelEncoder()
le2 = LabelEncoder()
le3 = LabelEncoder()
```

```python
X_train['sex'] = le1.fit_transform(X_train['sex'])
X_train['smoker'] = le2.fit_transform(X_train['smoker'])
X_train['time'] = le3.fit_transform(X_train['time'])
```

```python
X_train.head()
```

```python
X_test['sex'] = le1.transform(X_test['sex'])
X_test['smoker'] = le2.transform(X_test['smoker'])
X_test['time'] = le3.transform(X_test['time'])
```

```python
X_test.head()
```

```python
X_train['day'].value_counts()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer(transformers=[('onehot',OneHotEncoder(drop='first'),[3])], remainder='passthrough')
```

```python
ct.fit_transform(X_train)
```

```python
import sys
import numpy as np
np.set_printoptions(threshold=sys.maxsize)
X_train = ct.fit_transform(X_train)
```

```python
X_test = ct.transform(X_test)
```

```python
from sklearn.svm import SVR
svr = SVR()
```

```python
svr.fit(X_train, y_train)
```

```python
y_pred = svr.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
print(r2_score(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'C':[0.1,1,10,100,1000],
    'gamma':[1,0.1,0.01,0.001,0.0001],
    'kernel':['rbf']
}
```

```python
grid = GridSearchCV(SVR(), param_grid=param_grid, cv=5, refit=True, verbose=3)
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
gp = grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
print(r2_score(y_test, gp))
print(mean_absolute_error(y_test, gp))
```
